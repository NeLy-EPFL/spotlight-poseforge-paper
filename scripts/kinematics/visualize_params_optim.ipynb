```python
import polars as pl
import matplotlib.pyplot as plt
import optuna

from wc26.common.io import get_outputs_dir
```

```python
study_name = "decompFalse_multiobjTrue"
study_dir = (
    get_outputs_dir() / "neuromechfly_replay/data/nmf_phys_params_tuning" / study_name
)
trial_summary = pl.read_csv(study_dir / "trials_summary.csv")
db_url = f"sqlite:///{study_dir / 'optuna_study.db'}"
study = optuna.load_study(study_name=study_name, storage=db_url)
```

```python
trial_summary
```

```python
fig, ax = plt.subplots()
ax.scatter(trial_summary["values_0"], trial_summary["values_1"], s=1)
ax.set_xlabel("Trajectory alignment mismatch")
ax.set_ylabel("Trajectory ruggedness mismatch")
```

```python
import optuna.visualization

optuna.visualization.plot_pareto_front(
    study,
    target_names=["Trajectory alignment mismatch", "Trajectory ruggedness mismatch"],
)
```

```python
optuna.visualization.plot_param_importances(
    study, target=lambda t: t.values[0], target_name="Trajectory alignment mismatch"
)
```

```python
optuna.visualization.plot_param_importances(
    study, target=lambda t: t.values[1], target_name="Trajectory ruggedness mismatch"
)
```

```python
fig = optuna.visualization.plot_contour(
    study, target=lambda t: t.values[0], target_name="Trajectory alignment mismatch",
    params=["actuator_gain", "joint_damping", "sliding_friction"]
)
fig.update_layout(width=1600, height=1400)
```
